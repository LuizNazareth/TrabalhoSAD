```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
##lendo os datasets demograficos de 2016

insurance2016 = pd.read_excel('2016 - Insurance.xlsx')
schoolEnrollment2016 = pd.read_excel('2016 - School Enrollment.xlsx')
incomePerRace2016 = pd.read_excel('2016 - Income Per Race.xlsx')
race2016 = pd.read_excel('2016 - Race.xlsx')
poverty2016 = pd.read_excel('2016 - Poverty.xlsx')
```

```python
##dando merge nesses datasets

merge1 = pd.merge(insurance2016, schoolEnrollment2016, on='States')
merge1
```

```python
##dando merge nesses datasets

merge2 = pd.merge(merge1, incomePerRace2016, on='States')
merge2
```

```python
##dando merge nesses datasets

merge3 = pd.merge(merge2, race2016, on='States')
merge3
```

```python
##dando merge nesses datasets

data2016 = pd.merge(merge3, poverty2016, on='States')
data2016
```

```python
##alterando nome das colunas para Alabam2016 por ex

data2016['States'] = data2016['States'] + '2016'
data2016
```

```python
##colunas com objeto

colunas_percentuais = data2016.select_dtypes(include=['object']).columns
colunas_percentuais
```

```python
# Funo para remover o smbolo '%' e converter em float
def convert_percentage_to_float(value):
    if isinstance(value, str) and '%' in value:
        return float(value.replace('%', '')) / 100
    return value
```

```python
## colunas com float


cols_to_convert_percentage = data2016.columns.difference(['States'])

# Aplicar a funo para converter percentuais em floats
data2016[cols_to_convert_percentage] = data2016[cols_to_convert_percentage].applymap(convert_percentage_to_float)

# # Exibindo o DataFrame atualizado
# print(data2016)
print(data2016['Income per Race: Some other race'])
```

```python
def replace_comma_with_empty(value):
    if isinstance(value, str) and ',' in value:
        return value.replace(',', '')
    return value
```

```python
# Aplicar a funo a todo o DataFrame
data2016 = data2016.applymap(replace_comma_with_empty)

# Verificar o DataFrame aps a aplicao
data2016

```
