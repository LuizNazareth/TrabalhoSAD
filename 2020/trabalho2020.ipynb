```python
##lendo os datasets demograficos de 2020

insurance2020 = pd.read_excel('2020 - Insurance.xlsx')
schoolEnrollment2020 = pd.read_excel('2020 - School Enrollment.xlsx')
incomePerRace2020 = pd.read_excel('2020 - Income Per Race.xlsx')
race2020 = pd.read_excel('2020 - Race.xlsx')
poverty2020 = pd.read_excel('2020 - Poverty.xlsx')
```

```python
##dando merge nesses datasets

merge1 = pd.merge(insurance2020, schoolEnrollment2020, on='States')
merge1
```

```python
##dando merge nesses datasets

merge2 = pd.merge(merge1, incomePerRace2020, on='States')
merge2
```

```python
##dando merge nesses datasets

merge3 = pd.merge(merge2, race2020, on='States')
merge3
```

```python
##dando merge nesses datasets

data2020 = pd.merge(merge3, poverty2020, on='States')
data2020
```

```python
##alterando nome das colunas para Alabam2020 por ex

data2020['States'] = data2020['States'] + '2020'
data2020
```

```python
##colunas com objeto

colunas_percentuais = data2020.select_dtypes(include=['object']).columns
colunas_percentuais
```

```python
# Converter as colunas de 'Income per Race' diretamente para floats
cols_to_convert_directly = data2020.columns[data2020.columns.str.startswith('Income per Race')]
data2020[cols_to_convert_directly] = data2020[cols_to_convert_directly].apply(pd.to_numeric, errors='coerce')
```

```python
# Função para remover o símbolo '%' e converter em float
def convert_percentage_to_float(value):
    if isinstance(value, str) and '%' in value:
        return float(value.replace('%', '')) / 100
    return value
```

```python
## colunas com float

cols_to_convert_percentage = data2020.columns.difference(['States'] + list(cols_to_convert_directly))

# Aplicar a função para converter percentuais em floats
data2020[cols_to_convert_percentage] = data2020[cols_to_convert_percentage].applymap(convert_percentage_to_float)

# # Exibindo o DataFrame atualizado
# print(data2020)
```

```python
def replace_comma_with_empty(value):
    if isinstance(value, str) and ',' in value:
        return value.replace(',', '')
    return value
```

```python
# Aplicar a função a todo o DataFrame
data2020 = data2020.applymap(replace_comma_with_empty)

# Verificar o DataFrame após a aplicação
data2020

```
